```python
import numpy as np
import pandas as pd
```

```python
# read file train
train = pd.read_csv("train.csv")
train
```

```python
# read file test
test = pd.read_csv("test.csv")
test.tail()
```

```python
# fill missing value by median (numeric feature)
train["Age"] = train["Age"].fillna(train["Age"].median())
test["Age"] = test["Age"].fillna(train["Age"].median())
```

```python
# convert feature from category to numeric
train.loc[train["Embarked"] == 'S', "Embarked"] = 0.0
train.loc[train["Embarked"] == 'C', "Embarked"] = 1.0
train.loc[train["Embarked"] == 'Q', "Embarked"] = 2.0
train.loc[train["Sex"] == 'male', "Sex"] = 0.0
train.loc[train["Sex"] == 'female', "Sex"] = 1.0
test.loc[test["Embarked"] == 'S', "Embarked"] = 0.0
test.loc[test["Embarked"] == 'C', "Embarked"] = 1.0
test.loc[test["Embarked"] == 'Q', "Embarked"] = 2.0
test.loc[test["Sex"] == 'male', "Sex"] = 0.0
test.loc[test["Sex"] == 'female', "Sex"] = 1.0
```

```python
# drop missing value (category feature)
train.dropna(subset=['Embarked'], inplace=True)
test.dropna(subset=['Embarked'], inplace=True)
```

```python
# set input feature and target
x_train = np.array(train[["Pclass", "Sex", "Age", "Embarked"]].values, dtype=float)
y_train = np.array(train["Survived"].values, dtype=float)
x_test = np.array(test[["Pclass", "Sex", "Age", "Embarked"]].values, dtype=float)
```

```python
from Regression.LogisticRegression import LogisticRegression

# train logistic regression
model=LogisticRegression()
theta,bias,_=model.train(x_train,y_train,10000,0.00001,889)
```

```python
# weight of logistic regression
theta,bias
```

```python
# probability of test set
model.probability(x_test)
```

```python
# predict test set
# probability >= 0.5 -> predict 1
# probability < 0.5  -> predict 0
model.predict(x_test,0.5)
```
