```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read file csv
data = pd.read_csv("hr-employee-attrition-with-null.csv")
```

```python
# encoded target Yes,No -> 1.0,0.0
data.loc[data["Attrition"] == "No", "Attrition"] = 0.0
data.loc[data["Attrition"] == "Yes", "Attrition"] = 1.0
data = data.drop(columns= "EmployeeNumber")
data
```

```python
# encoded input feature (category to numeric)
for col in data.columns:
    data[col] = pd.Categorical(data[col]).codes
data.replace(-1, np.nan, inplace=True)
data
```

```python
# split data (test set)
data_neg = data.loc[data["Attrition"] == 0.0]
data_pos = data.loc[data["Attrition"] == 1.0]
test_neg = data_neg.sample(frac=0.1, random_state=42)
test_pos = data_pos.sample(frac=0.1, random_state=42)
test_set = pd.concat([test_pos, test_neg])
test_set = test_set.sort_index()
test_set
```

```python
# split data train set
test_indices = test_set.index
train_set = data.drop(test_indices)
train_set
```

```python
# set feature and target
x_train = train_set.drop(columns="Attrition")
y_train = train_set["Attrition"]
x_test = test_set.drop(columns="Attrition")
y_test = test_set["Attrition"]
```

```python
# compared histograms of features (Age, MonthlyIncome, DistanceFromHome)
# using number of bins (10,40,100)
for col in ["Age", "MonthlyIncome", "DistanceFromHome"]:
    print("feature:",col)
    for n_bin in [10,40,100]:
        print("bin:",n_bin)
        # remove NaN values
        x_train_col_no_nan = x_train[col][~np.isnan(train_set[col])]
        # bin the data into n_bin equally spaced bins
        # hist is the count for each bin
        # bin_edge is the edge values of the bins
        hist, bin_edge = np.histogram(x_train_col_no_nan, n_bin)
        # make sure to import matplotlib.pyplot as plt
        # plot the histogram
        plt.fill_between(bin_edge.repeat(2)[1:-1],hist.repeat(2))
        plt.show()
```

```python
# choose n_bins=10
n_bins=10
```

```python
# visualized likelihood distribution of features (MonthlyIncome, JobRole, HourlyRate, MaritalStatus)
# label : Yes -> green , No -> red
for col in ["MonthlyIncome", "JobRole", "HourlyRate", "MaritalStatus"]:
    print("feature:",col)
    Prob_pos = []
    Prob_neg = []
    Prob_between = []
    x_pos = x_train[col][y_train==1.0]
    x_neg = x_train[col][y_train==0.0]
    # remove NaN values
    x_train_col_no_nan = x_train[col][~np.isnan(x_train[col])]
    x_pos_no_nan = x_pos[~np.isnan(x_pos)]
    x_neg_no_nan = x_neg[~np.isnan(x_neg)]
    # bin the data into n_bin equally spaced bins
    # hist is the count for each bin
    # bin_edge is the edge values of the bins
    hist, bin_edge = np.histogram(x_train_col_no_nan, n_bins)
    pos_hist, pos_edges = np.histogram(x_pos_no_nan, bins=bin_edge)
    neg_hist, neg_edges = np.histogram(x_neg_no_nan, bins=bin_edge)

    for i in range(len(bin_edge)-1):
        epsilon = 1e-15
        prob_pos = pos_hist[i] / max(np.sum(pos_hist), epsilon)
        prob_neg = neg_hist[i] / max(np.sum(neg_hist), epsilon)

        # Setting epsilon to minimun probbabilities 
        prob_pos = max(prob_pos, epsilon)
        prob_neg = max(prob_neg, epsilon)

        Prob_pos.append(prob_pos)
        Prob_neg.append(prob_neg)
        Prob_between.append(min(prob_pos,prob_neg))
    
    # make sure to import matplotlib.pyplot as plt
    # plot the histogram
    plt.fill_between(bin_edge.repeat(2)[1:-1],np.array(Prob_pos).repeat(2),facecolor='mediumseagreen')
    plt.fill_between(bin_edge.repeat(2)[1:-1],np.array(Prob_neg).repeat(2),facecolor='lightcoral')
    plt.fill_between(bin_edge.repeat(2)[1:-1],np.array(Prob_between).repeat(2),facecolor='palegoldenrod')
    plt.legend(('Yes','No'),loc='upper right')
    plt.show()
```

```python
from Matrics.EvaluationMatrix import Evaluation
from BayesClassifier.SimpleBayesClassifier import SimpleBayesClassifier
from BayesClassifier.GaussianBayesClassifier import GaussianBayesClassifier
```

```python
# predict by SimpleBayesClassifier
model1=SimpleBayesClassifier()
model1.fit_params(x_train, y_train)
pred1=np.array(model1.predict(x_test))
print(pred1)
score=Evaluation(np.array(y_test),pred1)
print("Accuracy  :",score.accuracy())
print("Precision :",score.precision())
print("Recall    :",score.recall())
print("F1 score  :",score.F1())
```

```python
# predict by GuassianBayesClassifier
model2=GaussianBayesClassifier()
model2.fit_params(x_train, y_train)
pred2=np.array(model2.predict(x_test))
print(pred2)
score=Evaluation(np.array(y_test),pred2)
print("Accuracy  :",score.accuracy())
print("Precision :",score.precision())
print("Recall    :",score.recall())
print("F1 score  :",score.F1())
```

```python
# set baseline (random 50% -> 0.0, 50% -> 1.0)
pred0=np.random.randint(2, size=len(y_test))
print(pred0)
score=Evaluation(np.array(y_test),pred0)
print("Accuracy  :",score.accuracy())
print("Precision :",score.precision())
print("Recall    :",score.recall())
print("F1 score  :",score.F1())
```

```python
# The majority rule is the accuracy if you use the most frequent class
# from the training set as the classification decision. Report the overall Accuracy.
Mean=np.mean(y_train)
Most=0 if Mean < 0.5 else 1
predM=np.ones(len(y_test))*Most
print(predM)
score=Evaluation(np.array(y_test),predM)
print("Accuracy  :",score.accuracy())
print("Precision :",score.precision())
print("Recall    :",score.recall())
print("F1 score  :",score.F1())
```

```python
#  findtune the threshold that best F1 score (SimpleBayesClassifier)
thresh = np.arange(-5,5,0.05)
best_f1=0
for t in thresh:
    pred1=np.array(model1.predict(x_test, thresh=t))
    score=Evaluation(np.array(y_test),pred1)
    if score.F1() > best_f1:
        best_thresh = t
        best_f1 = score.F1()

pred1=np.array(model1.predict(x_test, thresh=best_thresh))
print(pred1)
print("threshold :", best_thresh)
score=Evaluation(np.array(y_test),pred1)
print("Accuracy  :",score.accuracy())
print("Precision :",score.precision())
print("Recall    :",score.recall())
print("F1 score  :",score.F1())
```

```python
#  findtune the threshold that best F1 score (GaussianBayesClassifier)
thresh = np.arange(-5,5,0.05)
best_f1=0
for t in thresh:
    pred2=np.array(model2.predict(x_test, thresh=t))
    score=Evaluation(np.array(y_test),pred2)
    if score.F1() > best_f1:
        best_thresh = t
        best_f1 = score.F1()

pred2=np.array(model2.predict(x_test, thresh=best_thresh))
print(pred2)
print("threshold :", best_thresh)
score=Evaluation(np.array(y_test),pred2)
print("Accuracy  :",score.accuracy())
print("Precision :",score.precision())
print("Recall    :",score.recall())
print("F1 score  :",score.F1())
```

```python
# findtune the threshold using ROC curve (SimpleBayesClassifier)
thresh = np.arange(-5,5,0.05)
true_pos_rate,false_pos_rate = [],[]
best_tpr, best_fpr = 0.0,1.0
for t in thresh:
    pred1=np.array(model1.predict(x_test, thresh=t))
    score=Evaluation(np.array(y_test),pred1)
    tp, fn, fp, tn = score.confusion_matrix()
    tpr,fpr = tp/max(tp+fn,1e-9),fp/max(fp+tn,1e-9)
    true_pos_rate.append(tpr)
    false_pos_rate.append(fpr)
    if tpr > best_tpr:
        best_thresh = t
        best_tpr = tpr
        best_fpr = fpr
    if tpr == best_tpr and fpr < best_fpr:
        best_thresh = t
        best_fpr = fpr

# Plot ROC curve
plt.figure()
plt.plot(false_pos_rate, true_pos_rate, color='darkorange', lw=2)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
# Plot best point on the ROC curve
plt.plot(best_fpr, best_tpr, 'ro') 
plt.show()


pred1=np.array(model1.predict(x_test, thresh=best_thresh))
print(pred1)
print("threshold :", best_thresh)
score=Evaluation(np.array(y_test),pred1)
print("Accuracy  :",score.accuracy())
print("Precision :",score.precision())
print("Recall    :",score.recall())
print("F1 score  :",score.F1())
```

```python
# shuffle data and calculate classifier mean and variance
model_score = {"acc": [], "pre" : [], "rec": [], "f1": []}
for _ in range(20):
    # split data (test set)
    test_neg = data_neg.sample(frac=0.1)
    test_pos = data_pos.sample(frac=0.1)
    test_set = pd.concat([test_pos, test_neg])
    test_set = test_set.sort_index()
    # split data train set
    test_indices = test_set.index
    train_set = data.drop(test_indices)
    # set feature and target
    x_train = train_set.drop(columns="Attrition")
    y_train = train_set["Attrition"]
    x_test = test_set.drop(columns="Attrition")
    y_test = test_set["Attrition"]
    # predict by SimpleBayesClassifier
    model1=SimpleBayesClassifier()
    model1.fit_params(x_train, y_train)
    pred1=np.array(model1.predict(x_test))
    score=Evaluation(np.array(y_test),pred1)
    model_score['acc'].append(score.accuracy())
    model_score['pre'].append(score.precision())
    model_score['rec'].append(score.recall())
    model_score['f1'].append(score.F1())

print("Accuracy  :", np.mean(model_score['acc']), np.std(model_score['acc']))
print("Precision :", np.mean(model_score['pre']), np.std(model_score['pre']))
print("Recall    :", np.mean(model_score['rec']), np.std(model_score['rec']))
print("F1 score  :", np.mean(model_score['f1']), np.std(model_score['f1']))
```
